## First merge

```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Create dataframes
path = r'C:\Users\rhtho\Documents\CareerFoundry\Data Immersion\Achievement 4\04-2023 Instacart Basket Analysis'
df_ords_clean_no_dups = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'df_ords_clean_no_dups.csv'))
df_prods_clean_no_dups = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'df_prods_clean_no_dups.csv'))
df_ords_prior = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'order_products__prior.csv'))
```

```python
df_dups = df_ords_prior[df_ords_prior.duplicated()]
```

```python
df_dups.shape
```

```python
# Merge dataframes
df_merged_large = df_ords_clean_no_dups.merge(df_ords_prior, on = 'order_id', indicator = True)
```
