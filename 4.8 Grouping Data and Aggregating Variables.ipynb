## Task

```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Create dataframes
ords_prods_merge = pd.read_pickle(os.path.join("D:\Data Immersion\orders_products_merged_4.7.pkl"))
```

```python
# Find aggregated mean
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

The results from the entire dataframe contain every department, whereas the results from the first million rows of the dataframe included only a handful of departments.  Additionally, the mean figures are different because the averages include many more order numbers for each department.  The department with the highest mean order number was 16 in the subset while it's department 21 in the full data set.

```python
# Create max order column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Create loyalty flags
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Find mean product prices for each loyalty group
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
# Create average price column
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Create spending flag
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge.head()
```

```python
# Create median days since prior order column
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Create order frequency flag
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna=False)
```
